```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

print("Setup complete")
```

```python
# init data frame

df = pd.DataFrame({'x':[79.31, 57.43, 60.66, 90.55, 92.12, 71.3, 70.5, 91.25], 
                   'y':[5.89, 3.84, 6.19, 9.22, 7.87, 6.29, 4.43, 8.91]})
display(df.T)
df = df.sort_values(by=['x'])
x = df.x
y = df.y
df.T
```

$ \hat{y}=b_1\cdot x+b_0$

```python
plt.grid()

plt.scatter(x, y, marker="o", label="Orgnl")

coef = np.polyfit(x, y, 1)
poly1d_fn = np.poly1d(coef)
plt.plot(x, poly1d_fn(x), label="Linear regresion", color="orange")
plt.legend()
print(f"LSM:{np.sum(np.power(y-poly1d_fn(x), 2))}")
```

$ x\cdot \hat{y}=b_1\cdot x + b_0$

```python
plt.grid()

plt.scatter(x, y, marker="o", label="Orgnl")

coef = np.polyfit(x, x*y, 1)
poly_hyper_d_fn = np.poly1d(coef)

plt.plot(x, poly_hyper_d_fn(x)/x, label="Hyperbolic regresion", color="orange")
plt.legend()
print(f"LSM:{np.sum(np.power(y-poly_hyper_d_fn(x)/x, 2))}")
```

$ \ln(\hat{y})=b_1\cdot \ln{x}+b_0 \implies \hat{y}=x^{b_1}\cdot e^{b_0}$

```python
plt.grid()

plt.scatter(x, y, marker="o", label="Orgnl")

coef = np.polyfit(np.log(x), np.log(y), 1)
poly_power_d_fn = np.poly1d(coef)

plt.plot(x, np.e**poly_power_d_fn(np.log(x)), label="Power regresion", color="orange")
plt.legend()
print(f"LSM:{np.sum(np.power(y-np.e**poly_power_d_fn(np.log(x)), 2))}")
```

$ \ln{y}=b_1\cdot x + b_0 \space \implies y=e^{b_1\cdot x + b_0}$

```python
plt.grid()

plt.scatter(x, y, marker="o", label="Orgnl")

coef = np.polyfit(x, np.log(y), 1)
poly_exp_d_fn = np.poly1d(coef)

plt.plot(x, np.e**poly_exp_d_fn(x), label="Exp regresion", color="orange")
plt.legend()
print(f"LSM:{np.sum(np.power(y-np.e**poly_exp_d_fn(x), 2))}")
print(f"y=e^({poly_exp_d_fn[0]}+({poly_exp_d_fn[1]}x))")
```

$ y=b_1\cdot \ln{x}+b_0$

```python
plt.grid()

plt.scatter(x, y, marker="o", label="Orgnl")

coef = np.polyfit(np.log(x), y, 1)
poly_log_d_fn = np.poly1d(coef)

plt.plot(x, poly_log_d_fn(np.log(x)), label="Log regresion", color="orange")
plt.legend()
print(f"LSM:{np.sum(np.power(y-poly_log_d_fn(np.log(x)), 2))}")
```
